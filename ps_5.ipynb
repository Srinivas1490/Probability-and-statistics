```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Part 1: Store Visit Patterns

def generate_store_data(N):
    np.random.seed(42)  # For reproducibility
    return np.random.randint(0, 11, N)

def plot_pmf_cdf(data):
    values, counts = np.unique(data, return_counts=True)
    pmf = counts / np.sum(counts)
    cdf = np.cumsum(pmf)

    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    # Plot PMF
    ax[0].bar(values, pmf, color='skyblue', edgecolor='black')
    ax[0].set_title("PMF: Probability of Visitors per Day")
    ax[0].set_xlabel("Number of Customers")
    ax[0].set_ylabel("Probability")

    # Plot CDF
    ax[1].plot(values, cdf, marker='o', linestyle='-', color='red')
    ax[1].set_title("CDF: Cumulative Probability of Visitors")
    ax[1].set_xlabel("Number of Customers")
    ax[1].set_ylabel("Cumulative Probability")

    plt.show()

# Generate and visualize store visit data
N_days = 100
store_data = generate_store_data(N_days)
plot_pmf_cdf(store_data)
```

```python
# Part 2: Exam Analysis

def binomial_exam_analysis(N, p, k):
    prob_k = stats.binom.pmf(k, N, p)
    prob_geq_75 = np.sum(stats.binom.pmf(np.arange(75, N+1), N, p))
    expected_value = N * p

    return prob_k, prob_geq_75, expected_value

# Exam analysis
N_students = 100
p_pass = 0.7
k_exact = 60
prob_60, prob_at_least_75, expected_pass = binomial_exam_analysis(N_students, p_pass, k_exact)

print(f"P(Exactly 60 students pass) = {prob_60:.4f}")
print(f"P(At least 75 students pass) = {prob_at_least_75:.4f}")
print(f"Expected students passing = {expected_pass}")
```
