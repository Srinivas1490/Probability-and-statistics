```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
mean = 75
std_dev = 10
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Histogram')
```

```python
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r', linewidth=2, label='PDF Curve')
```

```python
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution Histogram and PDF')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
mean_bmi = 20
std_dev_bmi = 4
```

```python
p_gte_23 = 1 - stats.norm.cdf(23, mean_bmi, std_dev_bmi)
p_lte_23 = stats.norm.cdf(23, mean_bmi, std_dev_bmi)
p_between_23_35 = stats.norm.cdf(35, mean_bmi, std_dev_bmi) - stats.norm.cdf(23, mean_bmi, std_dev_bmi)
```

```python
print(f"P(BMI >= 23): {p_gte_23:.4f}")
print(f"P(BMI <= 23): {p_lte_23:.4f}")
print(f"P(23 <= BMI <= 35): {p_between_23_35:.4f}")
```

```python
x_bmi = np.linspace(mean_bmi - 4*std_dev_bmi, mean_bmi + 4*std_dev_bmi, 1000)
pdf_bmi = stats.norm.pdf(x_bmi, mean_bmi, std_dev_bmi)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_bmi, pdf_bmi, 'r', linewidth=2, label='BMI PDF')
plt.fill_between(x_bmi, pdf_bmi, where=(x_bmi >= 23), color='blue', alpha=0.3, label='BMI >= 23')
plt.fill_between(x_bmi, pdf_bmi, where=(x_bmi <= 23), color='green', alpha=0.3, label='BMI <= 23')
plt.fill_between(x_bmi, pdf_bmi, where=((x_bmi >= 23) & (x_bmi <= 35)), color='purple', alpha=0.3, label='23 <= BMI <= 35')
```

```python
plt.xlabel('BMI')
plt.ylabel('Density')
plt.title('BMI Distribution and Probabilities')
plt.legend()
plt.grid()
plt.show()
```
