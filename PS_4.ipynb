```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/content/employee_data_250-2.csv"
df = pd.read_csv(file_path)

sns.set_style("whitegrid")

def plot_salary_histogram():
    plt.figure(figsize=(10, 6))
    sns.histplot(df["Salary"], bins=15, kde=True, color="blue")
    plt.xlabel("Salary (USD)", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.title("Distribution of Employee Salaries", fontsize=14)
    plt.show()

def plot_experience_vs_salary():
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df["Experience (Years)"], y=df["Salary"], alpha=0.6)
    sns.regplot(x=df["Experience (Years)"], y=df["Salary"], scatter=False, color="red")
    plt.xlabel("Experience (Years)", fontsize=12)
    plt.ylabel("Salary (USD)", fontsize=12)
    plt.title("Experience vs. Salary", fontsize=14)
    plt.show()

def plot_department_distribution():
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df["Department"], order=df["Department"].value_counts().index, palette="viridis")
    plt.xlabel("Number of Employees", fontsize=12)
    plt.ylabel("Department", fontsize=12)
    plt.title("Number of Employees in Each Department", fontsize=14)
    plt.show()

plot_salary_histogram()
plot_experience_vs_salary()
plot_department_distribution()
```
